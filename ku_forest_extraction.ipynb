# Convert String to JSON

```python
path_file = '/content/drive/MyDrive/KnowledgeGraph/Json_researcher/ku-forest.json'
```

```python
f = open(path_file,"r",encoding='utf8')
```

```python
f = open(path_file, "r",encoding="utf8")
count = 0
content = []
string_json = ''
begin = '['
end = ']'
split = ','

string_json += begin
while(True):
    # Read a line.
    line = f.readline()
    # When readline returns an empty string, the file is fully read.
    if line == "":
        print("::DONE::")
        break
    if count != 0:
      string_json += split
    string_json += line
    content.append(line)
    count +=1

string_json += end
```

```python
import json
data = json.loads(string_json)
```

```python
df_publish = pd.json_normalize(data)
```
